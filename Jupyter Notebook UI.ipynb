# Mode 1

```python
import actionlib
from move_base_msgs.msg import MoveBaseAction
from ipywidgets import interact

from geometry_msgs.msg import Twist
pubv = rospy.Publisher('/cmd_vel', Twist, queue_size=1)

global X, Y

# Action client
ac = actionlib.SimpleActionClient('move_base', MoveBaseAction)
ac.wait_for_server()

goal = MoveBaseGoal()

def SendGoal(x,y):
    print ('Choose the (x,y) position\n')
    goal.target_pose.header.frame_id = "map"
    goal.target_pose.header.stamp = rospy.Time.now()
    goal.target_pose.pose.orientation.w = 1.0
    goal.target_pose.pose.position.x = x
    goal.target_pose.pose.position.y = y
    ac.send_goal(goal)
    
X = widgets.IntSlider(min=-20, max=20, step=1, value=0, description="x")
Y = widgets.IntSlider(min=-20, max=20, step=1, value=0, description="y")
    
interact(SendGoal, x=X, y=Y);

CG = Button(description='Cancel Goal',
                        layout=Layout(width='auto', grid_area='C'),
                        style=ButtonStyle(button_color='salmon'))
NG = Button(description='New Goal',
                        layout=Layout(width='auto', grid_area='N'),
                        style=ButtonStyle(button_color='lightgreen'))

def NewGoal(N):
    print('Choose a new goal\n')
    SendGoal(x,y)
    
def CancelGoal(C):
    vel = Twist()
    print('Goal cancelled\n')
    vel.linear.x = 0
    vel.angular.z = 0
    pubv.publish(vel)


Box = HBox([CG, NG])
display(X,Y,B)

NG.on_click(NewGoal)
CG.on_click(CancelGoal)
```

# Mode 2

```python
from geometry_msgs.msg import Twist
pubv = rospy.Publisher('/cmd_vel', Twist, queue_size=1)

# Teleoperation buttons
F = Button(description='Go Forward',
                        layout=Layout(width='auto', grid_area='F'),
                        style=ButtonStyle(button_color='white'))
B = Button(description='Go Back',
                        layout=Layout(width='auto', grid_area='B'),
                        style=ButtonStyle(button_color='white'))
L = Button(description='Turn Left',
                        layout=Layout(width='auto', grid_area='L'),
                        style=ButtonStyle(button_color='white'))
R = Button(description='Turn Right',
                        layout=Layout(width='auto', grid_area='R'),
                        style=ButtonStyle(button_color='white'))
FL = Button(description='Go Straight-Left',
                        layout=Layout(width='auto', grid_area='FL'),
                        style=ButtonStyle(button_color='white'))
FR = Button(description='Go Straight-Right',
                        layout=Layout(width='auto', grid_area='FR'),
                        style=ButtonStyle(button_color='white'))
BL = Button(description='Go Back-Left',
                        layout=Layout(width='auto', grid_area='BL'),
                        style=ButtonStyle(button_color='white'))
BR = Button(description='Go Back-Right',
                        layout=Layout(width='auto', grid_area='BR'),
                        style=ButtonStyle(button_color='white'))

Inc = Button(description='Increase Velocity',
                        layout=Layout(width='auto', grid_area='IV'),
                        style=ButtonStyle(button_color='lightgreen'))
Res = Button(description='Reset Velocity',
                        layout=Layout(width='auto', grid_area='RV'),
                        style=ButtonStyle(button_color='salmon'))
S = Button(description='Stop',
                        layout=Layout(width='auto', grid_area='S'),
                        style=ButtonStyle(button_color='red'))


def GoForward(b):
    vel = Twist()
    speed = 0.5
    vel.linear.x = 1.0*speed
    vel.angular.z = 0
    pubv.publish(vel)
    
def GoBack(b):
    vel = Twist()
    speed = 0.5
    vel.linear.x = -1.0*speed
    vel.angular.z = 0
    pubv.publish(vel)
    
def TurnLeft(b):
    turn = 1.0
    vel = Twist()
    vel.linear.x = 0
    vel.angular.z = 1.0*turn
    pubv.publish(vel)
        
def TurnRight(b):
    turn = 1
    vel = Twist()
    vel.linear.x = 0
    vel.angular.z = -1*turn
    pubv.publish(vel)
        
def GoStraightLeft(b):
    speed = 0.5
    turn = 1
    vel = Twist()
    vel.linear.x = 1.0*speed
    vel.angular.z = 1.0*turn
    pubv.publish(vel)
        
def GoStraightRight(b):
    speed = 0.5
    turn = 1
    vel = Twist()
    vel.linear.x = 1*speed
    vel.angular.z = -1*turn
    pubv.publish(vel)
    
def GoBackLeft(b):
    speed = 0.5
    turn = 1.0
    vel = Twist()
    vel.linear.x = -1.0*speed
    vel.angular.z = 1.0*turn
    pubv.publish(vel)

def GoBackRight(b):
    speed = 0.5
    turn = 1
    vel = Twist()
    vel.linear.x = -1.0*speed
    vel.angular.z = -1.0*turn
    pubv.publish(vel)
    
def IncreaseVelocity(b):
    speed = 0.5
    turn = 1
    speed = speed+1.0
    turn = turn+0.5
    print('Lienar Velocity: ', speed)
    print('Angula Velocity: ', turn)
    pubv.publish(vel)
    
def ResetVelocity(b):
    speed = 0.5
    turn = 1.0
    pubv.publish(vel)
    
def Stop(b):
    speed = 0
    turn = 0
    pubv.publish(vel)
    
F.on_click(GoForward)
B.on_click(GoBack)
L.on_click(TurnLeft)
R.on_click(TurnRight)
FL.on_click(GoStraightLeft)
FR.on_click(GoStraightRight)
BL.on_click(GoBackLeft)
BR.on_click(GoBackRight)
Inc.on_click(IncreaseVelocity)
Res.on_click(ResetVelocity)
S.on_click(Stop)

Box1 = HBox([ VBox([FL, L, BL]), VBox([F, S, B]), VBox([FR, R, BR]) ])
Box2 = HBox([Increase, Reset])
display(Box1)
display(Box2)
```
